for csv data

```python
from langchain_community.document_loaders import CSVLoader

loader = CSVLoader(
    file_path="patient_records.csv"
)
docs = loader.load()  #docs is a list of Document objects
len(docs)
```

```python
print(docs[0].page_content)
```

for url data

```python
from langchain_community.document_loaders import UnstructuredURLLoader

# pip install unstructured

loader = UnstructuredURLLoader(
    urls=["https://www.cnbc.com/2024/12/21/how-the-federal-reserves-rate-policy-affects-mortgages.html"]
)
docs = loader.load()
len(docs)
```

```python
docs[0].page_content
```

```python
from langchain_community.document_loaders import WebBaseLoader

loader = WebBaseLoader(
    web_paths=["https://www.cnbc.com/2024/12/21/how-the-federal-reserves-rate-policy-affects-mortgages.html"],
    header_template={
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 Chrome/120 Safari/537.36"
    }
)

docs = loader.load()
print(len(docs))
print(docs[0].page_content[:500])
```
